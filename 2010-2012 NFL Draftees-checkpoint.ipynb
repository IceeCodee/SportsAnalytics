<h1> 2010-2012 NFL Draft Draftees </h1>

<p> The purpose of this notebook is to analyze NFL draft data from years 2010-2012 to find the schools who have the most number of draftees. </p>

```python
import pandas as pd
import nfl_data_py as nfl
# may need to install data using line below
#!pip install nfl_data_py
```

```python
#import data from 2010-2012
draft_2010 = nfl.import_combine_data([2010])
draft_2011 = nfl.import_combine_data([2011])
draft_2012 = nfl.import_combine_data([2012])

#draft_2010.head()
```

```python
# drop all NaN values, as these are the players who went undrafted
draft_2010.dropna()
draft_2011.dropna()
draft_2012.dropna()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#count the number of draftees for each school, only return top 5
count_2010 = draft_2010['school'].value_counts().head(5)
count_2011 = draft_2011['school'].value_counts().head(5)
count_2012 = draft_2012['school'].value_counts().head(5)
```

```python
data_2010 = count_2010.to_frame().reset_index()
schools_2010 = data_2010.rename(columns={'index':'school', 'school':'count'})
```

```python
plt.bar(schools_2010['school'], schools_2010['count'], color='purple')
plt.xlabel('Top 5 Schools')
plt.ylabel('Number of Players')
plt.title("Top Schools of 2010 NFL Draftees")
plt.show()
```

```python
data_2011 = count_2011.to_frame().reset_index()
schools_2011 = data_2011.rename(columns={'index':'school', 'school':'count'})
```

```python
plt.bar(schools_2011['school'], schools_2011['count'])
plt.xlabel('Top 5 Schools')
plt.ylabel('Number of Players')
plt.title("Top Schools of 2011 NFL Draftees")
plt.show()
```

```python
data_2012 = count_2012.to_frame().reset_index()
schools_2012 = data_2012.rename(columns={'index':'school', 'school':'count'})
```

```python
plt.bar(schools_2012['school'], schools_2012['count'], color='red')
plt.xlabel('Top 5 Schools')
plt.ylabel('Number of Players')
plt.title("Top Schools of 2012 NFL Draftees")
plt.show()
```

<h1>NFL Combine Data From 2022</h1>

<p> The puropose of this notebook is to compare metrics of the different positions from the 2022 combine. This notebook sources data from the nfl_data_py python package. More infomation on this library can be found here: https://github.com/cooperdff/nfl_data_py .</p>

```python
import nfl_data_py as nfl
import pandas as pd
import matplotlib.pyplot as plt
```

```python
combine_data = nfl.import_combine_data([2022])
combine_data.head()
```

<p> We only want to collect data from the 2022 combine for the following positons: quarter back, running back, wide reciever, tight end and safety </p>

```python
filter_positions = combine_data[(combine_data['pos'].isin(['S','WR','QB','RB','TE'])) & (combine_data['draft_year']==2022)]
#filter_positions.head()
```

<p> Forty yard dash box plot for each position. The QB positions has the slowest average time followed by the tight end position. Generally, the saftey position has the fastest average dash time </p>

```python
filter_positions.boxplot(by='pos', column=['forty'], patch_artist=True)
plt.title('Forty Yard Dash By Position')
plt.xlabel('Positions')
plt.ylabel('Seconds')
plt.suptitle('') #removes default title for chart
plt.show()
```

<p> Next is the bench press metric where each person lifts 225 lbs as many times as possible. It should be noted the wide recievers and quarter backs do not have data in this particalar data set. Per the boc plot below, running backs usually knock out the most reps. </p>

```python
filter_positions.boxplot(by='pos', column=['bench'], patch_artist=True)
plt.title('Bench Press Dash By Position')
plt.xlabel('Positions')
plt.ylabel('Reps')
plt.suptitle('') #removes default title for chart
plt.show()
```

<p> The vertical jump is an exercise where players jump as high straight up into the air as they can. Athletes in the safety position tends to have the highest average jump. Running backs have the biggest range in jumps and conversely, tight ends have low variablity (not including outliers). </p>

```python
filter_positions.boxplot(by='pos', column=['vertical'], patch_artist=True)
plt.title('Vertical Jump By Position')
plt.xlabel('Positions')
plt.ylabel('Inches')
plt.suptitle('') #removes default title for chart
plt.show()
```

<p>Broad jump is when atheletes perform a standing jump as far of a distance as possible. Likewise, with the broad jump tight ends have low variability, and quarter backs are at the low end of the spectrum. Positions wide reciever and saftey have the longest average jump with recievers bring centimeters ahead of RBs </p>

```python
filter_positions.boxplot(by='pos', column=['broad_jump'], patch_artist=True)
plt.title('Broad Jump By Position')
plt.xlabel('Positions')
plt.ylabel('Inches')
plt.suptitle('') #removes default title for chart
plt.show()
```

```python
import pandas as pd
import nfl_data_py as nfl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
weekly_data = nfl.import_weekly_data([2022])
QB_weekly_data = weekly_data[weekly_data['position'] == 'QB']
WR_weekly_data = weekly_data[weekly_data['position']=='WR']
RB_weekly_data = weekly_data[weekly_data['position']=='RB']
```

```python
QB_weekly_data['player_name'].unique()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.ylabel('Passing Yards')
plt.xlabel('Week')
plt.xticks(range(1,19))
ax.legend()
for qb_name in QB_weekly_data['player_name'].unique():
    # Filter data for the current QB
    qb_data = QB_weekly_data[QB_weekly_data['player_name'] == qb_name]
    
    # Plot passing yards data for the QB
    ax.plot(qb_data['week'], qb_data['passing_yards'], label=qb_name)
plt.show()
```
